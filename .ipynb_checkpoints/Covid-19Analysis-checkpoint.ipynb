```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"Corona.csv",index_col='Sr.No')
```

<h2 style="color:blue">Confirmed Status</h2>
<h2 style="color:red">Deceased Status</h2>
<h2 style="color:Green">Recovered Status</h2>

```python
df.head()
```

```python
data=df.drop(['Unnamed: 0'],axis=1)
```

```python
from pandas_profiling import ProfileReport
```

```python
profile=ProfileReport(data,title='Pandas Profiling Report', html={'style':{'full_width':True}} )
```

```python
profile.to_notebook_iframe()
```

```python
profile.to_file('pandasProfiling.html')
```

```python
plt.figure(figsize=(18,10))
plt.plot(data['States/UT'],data['Confirmed'])
plt.xticks(rotation=90)
plt.xlabel('State/UT')
plt.ylabel('Confirmed')
plt.title('Confirmed cases to covid-19 Statewise')
plt.show()
```

<h2 style="color:Blue">Confirmed Case Status</h2>

```python
df_max = data.groupby(['States/UT']).max().sort_values(by='Confirmed',ascending=False)
df_max.head(5)
```

<h2 style="color:red">Deceased Status</h2>

```python
df_max = data.groupby(['States/UT']).max().sort_values(by='Deceased',ascending=False)
df_max.head(5)
```

<h2 style="color:green">Recovering Status</h2>

```python
df_max = data.groupby(['States/UT']).max().sort_values(by='Recovered',ascending=False)
df_max.head(5)
```

<h2 style="color:darkgreen" >These States/UT Have No Death Due to Covid-19 Yet </h2>

```python
data.loc[df['Deceased']==0,['States/UT','Deceased']]
```

<h2 style="color:red" >These States/Ut has Confirm Case over 500 </h2>

```python
data.loc[df['Confirmed']>500,['States/UT','Confirmed']]
```
